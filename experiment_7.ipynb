```python
# Fibonacci Series using while loop
'''a = 0
b = 1
c = a + b
a = b
b = c'''
a = 0
b = 1
n = int(input("Enter number of desired iterations: "))

while n > 0:
  print(a, end=" ")
  c = a + b
  a = b
  b = c
  n -= 1
```

## The break Statement
### With the break statement we can stop the loop even if the while condition is true

```python
#Exit the loop when i is 3
i = 1
while i < 6:
  print(i)
  if i == 3:
    break
  i += 1
```

```python
#Python program to accept three digits and print all possible combinations
p1 = 1
p2 = 2
p3 = 3
p = [p1,p2,p3]
# 123,132,213,231,312,321
for i in range(0,3):
  for j in range(0,3):
    for k in range(0,3):
      if (p[i]!=p[j] and p[j]!=p[k] and p[k]!=p[i]):
        print(p[i],p[j],p[k])
```

```python
# Armstrong Number (153 370 371 407 123)
# sum of each digit raised to number of digits is equal to the number (1^3 + 5^3 + 3^3 = 1+ 125+ 9)
x = int(input("Enter a number: "))
total = 0

for digit in str(x):    ## int datatype is not iterable while str is iterable
  total += int(digit)**len(str(x))

if total == x:
  print("Provided number is an Armstrong Number!")
else:
  print("Try Again! Not an Armstrong Number!!")
```
